```python
import pandas as pd
total_data = pd.read_csv(r"https://raw.githubusercontent.com/4GeeksAcademy/k-means-project-tutorial/main/housing.csv")
total_data.to_csv(r"/workspaces/clustering-knn-model/data/raw/total_data")
```

```python
total_data
```

```python
total_data.drop(["HouseAge", "AveRooms", "AveBedrms", "Population","AveOccup","MedHouseVal",], axis = 1, inplace = True)
total_data
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
fig, axis = plt.subplots(2, 1, figsize = (10, 11))

# Crear una figura múltiple con histogramas y diagramas de caja
sns.histplot(ax = axis[0], data = total_data, x = "MedInc").set(xlabel = None)
sns.boxplot(ax = axis[1], data = total_data, x = "MedInc")
```

```python
from sklearn.model_selection import train_test_split

X = total_data
X_train, X_test = train_test_split(X, test_size=0.2, random_state=14)
```

```python
X_train.to_csv(r"/workspaces/clustering-knn-model/data/processed/X_train.csv", index=False)
X_test.to_csv(r"/workspaces/clustering-knn-model/data/processed/X_test.csv", index=False)
```
